```python
import os

lst = []
def read_skeleton(filename, person_id):
    # 讀取 .skeleton 檔
    with open(filename, 'r') as f:
        lines = f.readlines()
    # 第 1 行為該樣本的幀數
    num_frames = int(lines[0])
    # 第 2 行為執行動作的人數
    num_people = int(lines[1])
#     print(person_ID[0])
    frame_lines = []
    # 將每一幀的 Y 軸資料抓出來
    if num_people == 1:
    # 只有一個人，讀取所有行
        lines = lines[4:]
    else:
        lines = lines[4:]
        frame_lines = lines[:25]
        frame_lines = lines[25+25+3:] 
#     print(lines)
    y_data = []
    frame_data = []
       
    i = 0

    # 取出每一幀的 Y 軸資料
    for i in range(num_frames):
        # 每一幀有 25 個關節，每個關節有 1 個 Y 軸資料
        # 將所有 Y 軸資料抓出來
        # 取出第 i 幀的資料    
        if num_people == 1:
            for line in lines[i*25+3*i:i*25+(3*i)+25]:
                frame_data.append(float(line.split()[1]))
#                 print(frame_data)
            y_data.append(frame_data)
            frame_data = []
            i += 1
        else:
            limit = num_frames
            for q in range(num_frames):
                for line in lines[i*25+(30*q):i*25+25+(30*q)]:
                        frame_data.append(float(line.split()[1]))
#                         print(frame_data)
                        if q >= 1:
                            q = 0
                        else:
                            q += 1 
#                         print(q)
                y_data.append(frame_data)
                frame_data = []
                i += 1
            if i >= limit:
                break
#            print(y_data)
    
    # 計算第一幀與中間幀和最後一幀的變化量
    # 取 25 個點的平均值
    first_frame = y_data[0]
    mid_frame = y_data[num_frames//2]
    last_frame = y_data[-1]

#     print(first_frame[2])
#     print(mid_frame[2])
#     print(last_frame[2])
    avg_change_right_hand = (((first_frame[23]- mid_frame[23])) + ((mid_frame[23] - last_frame[23]))-
                             ((first_frame[10]- mid_frame[10])) + ((mid_frame[10] - last_frame[10])))/2 
    #僅計算關節24-11
    avg_change_left_hand = (((first_frame[21]- mid_frame[21])) + ((mid_frame[21] - last_frame[21]))-
                            ((first_frame[6]- mid_frame[6])) + ((mid_frame[6] - last_frame[6])))/2 
    #僅計算關節22-7
    avg_change_right_raise = (((mid_frame[21])-( mid_frame[20])))
    #僅計算關節22-21
    avg_change_left_raise = (((mid_frame[23])-( mid_frame[20])))
    #僅計算關節24-21
    avg_change_right_foot = (first_frame[13]- first_frame[19]) - (mid_frame[13]- mid_frame[19])
    #僅計算關節20-14
    avg_change_left_right = (first_frame[15]- mid_frame[15]) - (first_frame[19]- mid_frame[19])
    #僅計算關節16-20
    avg_change_kick =  abs((mid_frame[15]- mid_frame[19])) - abs((last_frame[15]- last_frame[19]))
    #僅計算關節16-20
    avg_change_left_foot = (first_frame[15]- first_frame[17] ) - (mid_frame[15]- mid_frame[17])
    #僅計算關節16-18
    avg_change_all = sum([y1-y2 for y1, y2 in zip(first_frame, mid_frame)]) / 25 + \
                 sum([y1-y2 for y1, y2 in zip(mid_frame, last_frame)]) / 25
    avg_change_all /= 2

    #25關節全參考
    
    lst.append(avg_change_right_hand)
    lst.append(avg_change_left_hand)
    lst.append(avg_change_right_foot)
    lst.append(avg_change_left_foot)
    lst.append(avg_change_all)
    lst.append(avg_change_right_raise)
    lst.append(avg_change_left_raise)
    lst.append(avg_change_left_right)
    lst.append(avg_change_kick)


# 讀取多個 .skeleton 檔
filenames = ['./0112_test/1/S001C001P004R001A023.skeleton']
person_ID = 1
for filename in filenames: 
    avg_change_3 = read_skeleton(filename, person_ID)
    avg_change_all = read_skeleton(filename, person_ID)
    print(f'文件 {filename} 的右手變化量為 {lst[0]:.3f}')
    print(f'文件 {filename} 的左手變化量為 {lst[1]:.3f}')
    print(f'文件 {filename} 的右腳變化量為 {lst[2]:.3f}')
    print(f'文件 {filename} 的左腳變化量為 {lst[3]:.3f}')
    print(f'文件 {filename} 的平均變化量為 {lst[4]:.3f}')
    print(f'文件 {filename} 的右手舉起變化量為 {lst[5]:.3f}')
    print(f'文件 {filename} 的左手舉起變化量為 {lst[6]:.3f}')
    print(f'文件 {filename} 的左腳與右腳變化量為 {lst[7]:.3f}')
    print(f'文件 {filename} 的第一幀與中間幀腳關節變化量為 {lst[8]:.3f}')


if  ((lst[0] > 0 or lst[1] > 0)  and (lst[5] > 0 or lst[6] > 0)) or ((lst[5] > 0.1 or lst[6] > 0.1)) :
    print('wave')
elif (abs(lst[2]) > 0.05 or lst[3] > 0.02) and (lst[7] > 0.03 or lst[8] > 0.01) and lst[4] < 0.05:
    print('kick')
elif lst[4] < 0 or lst[4] > 0.05:
    print('stand')
else:
    print('sit')
```
